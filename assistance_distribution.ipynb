```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset from CSV file
df = pd.read_csv('assistance_preference.csv')

# Preview all the data
df.head()
```

```python
# Count preferences
preference_counts = df['Preferred Assistance'].value_counts()

# Display preference counts
print("\nPreference Counts:\n", preference_counts)

```

```python
# Percentage of preferences
preference_percentage = preference_counts / preference_counts.sum() * 100
print("\nPreference Percentages:\n", preference_percentage)
```

```python
# Vizualize using Bar chart for assistance preferences
plt.bar(preference_counts.index, preference_counts.values, color=['blue', 'green', 'orange'])
plt.title('Assistance Preferences')
plt.xlabel('Assistance Type')
plt.ylabel('Number of Participants')
plt.show()
```

```python
# Vizualize using Pie chart for percentage contribution
plt.pie(preference_counts, labels=preference_counts.index, autopct='%1.1f%%', colors=['blue', 'green', 'orange'])
plt.title('Percentage of Assistance Preferences')
plt.show()
```
