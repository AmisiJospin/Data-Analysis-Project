```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset from CSV file
df = pd.read_csv('durable_solution_nationality.csv')

# Preview data
df.head()
```

```python
# Count the numner of preferences overall
solution_counts = df['Preferred Solution'].value_counts()

# Count preferences by nationality
nationality_solution_counts = df.groupby(['Nationality', 'Preferred Solution']).size().unstack(fill_value=0)

# Display the total counts
print("\nOverall Preference Counts:\n", solution_counts)
```

```python
# Display each counts
print("\nPreferences by Nationality:\n", nationality_solution_counts)
```

```python
# Vizualize using Bar chart for overall solution preferences
plt.bar(solution_counts.index, solution_counts.values, color=['cyan', 'chocolate', 'pink'])
plt.title('Overall Durable Solution Preferences')
plt.xlabel('Preferred Solution')
plt.ylabel('Number of Participants')
plt.show()
```

```python
# Visualize using bar chart for preferences by nationality
nationality_solution_counts.plot(kind='bar', stacked=True, figsize=(8, 6), color=['cyan', 'chocolate', 'pink'])
plt.title('Durable Solution Preferences by Nationality')
plt.xlabel('Nationality')
plt.ylabel('Number of Participants')
plt.legend(title='Preferred Solution')
plt.show()
```

```python
# Visualize using Pie chart for overall preferences
plt.pie(solution_counts, labels=solution_counts.index, autopct='%1.1f%%', colors=['cyan', 'chocolate', 'pink'])
plt.title('Percentage of Durable Solution Preferences')
plt.show()
```
