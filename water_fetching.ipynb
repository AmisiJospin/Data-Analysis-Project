```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset from CSV file
df = pd.read_csv('water_fetching_data.csv')

# Preview the data 
df.head()
```

```python
# Add a column for weekly containers
df['Weekly Containers'] = df['Containers Per Day'] * df['Day per Week']

# Add a column for monthly containers (approx. 4 weeks per month)
df['Monthly Containers'] = df['Weekly Containers'] * 4

# Display updated data
print("\nData with Weekly and Monthly Containers:\n", df)
```

```python
# Total containers that can be fetched per month in an area
total_containers = df[['Area', 'Monthly Containers']]
print("\nTotal Monthly Containers by Area:\n", total_containers)

```

```python
# Find average containers fetched per day
average_containers = df['Containers Per Day'].mean()
print(f"\nAverage Containers Fetched Per Day: {average_containers:.2f}")
```

```python
# Visualize using bar chart for monthly containers
plt.bar(df['Area'], df['Monthly Containers'], color='maroon')
plt.title('Monthly Water Containers Fetched by Area')
plt.xlabel('Area')
plt.ylabel('Monthly Containers')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize using Pie chart for contribution by area
plt.pie(df['Monthly Containers'], labels=df['Area'], autopct='%1.1f%%', colors=['maroon', 'green', 'red', 'orange', 'purple', 'cyan', 'magenta', 'yellow'])
plt.title('Percentage Contribution by Area')
plt.show()
```
